```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
boston
```

```python
feature_names = boston["feature_names"]
target = boston["target"]
data = boston["data"]
X = pd.DataFrame(data, columns=feature_names)
y = pd.DataFrame(target, columns=["price"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
r2_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators = 1000, max_depth = 12, random_state = 42)
```

```python
rf.fit(X_train, y_train.values[:, 0]);
```

```python
y_pred_rf = rf.predict(X_test)
y_pred_rf
```

```python
r2_score(y_test, y_pred)
```

```python
r2_score(y_test, y_pred_rf)
```

```python
X = df
X.drop(["Class"],axis=1)
```

```python
y = pd.Series(df['Class'].values)
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
y_test.size
```

```python
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
```

```python
clf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=100),
    param_grid=parameters,
    scoring='roc_auc',
    cv=3,
)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
predicted = clf.predict_proba(X_test)
predicted
```

```python
y_pred_proba = predicted[:,1]
```

```python
y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score
r2_score(y_test, y_pred_proba)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (15,10))
sns.set(font_scale=1.4)
X_corr = X.corr().abs()
sns.heatmap(X_corr, annot=True, linewidths=.5, cmap='coolwarm')
plt.title('Correlation matrix')
plt.show()
```

```python
X_corr_with = X_corr.corrwith(X_corr["target"])
```

```python
X_corr_with = X_corr_with.drop("target")
```

```python
X_corr_with = X_corr_with.abs()
X_corr_with
```

```python
X_corr_with = X_corr_with.sort_values(ascending=False)
X_corr_with
```

```python
high_corr = X_corr_with[X_corr_with>0.5].keys()
high_corr
```

```python
X_mod = X.drop(["target"],axis=1)
X_mod
```

```python
for high_corr_label in high_corr:
    X[high_corr_label+"_2"] = X[high_corr_label]**2
X
```
